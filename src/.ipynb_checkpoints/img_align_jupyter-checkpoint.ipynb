```python
from astropy.io import fits
import matplotlib.pyplot as plt
import aplpy
import os

path = os.getcwd()

name = 'M34_0001_slice_1'
gc = aplpy.FITSFigure(os.path.dirname(path)+f'/output/M34_{name}', north=True, figsize=(10,10))

gc.show_colorscale(cmap='inferno')
gc.add_grid()
gc.grid.set_alpha(0.5)
gc.ticks.set_minor_frequency(6)

gc.axis_labels.set_xtext('Right Ascension (J2000)')
gc.axis_labels.set_ytext('Declination (J2000)')
gc.axis_labels.set_font(size='xx-large')
gc.tick_labels.set_font(size='xx-large')
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.visualization import ZScaleInterval, imshow_norm, ImageNormalize
import os


file = fits.open(os.getcwd()+'/slices_M34_0001/slice_M34_0001_1.fits')
img = file[0].data


# norm
intervalo = ZScaleInterval()
vmin, vmax = intervalo.get_limits(img)
normalizador = ImageNormalize(vmin=vmin, vmax=vmax)

# vis
plt.figure(figsize=(8, 8))
plt.imshow(img, cmap='gray', origin='lower', norm=normalizador)
plt.colorbar(label='Unidades')
plt.show()
```
